## Consultas SQL con DuckDB

```python
import duckdb

duckdb.register("superstore", df)
```

Consulta 1. Ventas totales por categoría

```python
duckdb.sql("""SELECT
    Category,
    ROUND(SUM(Sales),2) AS Total_Ventas
FROM superstore
GROUP BY Category
ORDER BY Total_Ventas DESC;""")
```

Consulta 2. Utilidad promedio por región

```python
duckdb.sql("""SELECT
    Region,
    ROUND(AVG(Profit),2) AS Utilidad_Promedio
FROM superstore
GROUP BY Region
ORDER BY Utilidad_Promedio DESC;""")
```

Consulta 3. Top 10 productos con mayores ventas

```python
duckdb.sql("""SELECT
    Product_Name,
    SUM(Sales) AS Ventas
FROM (
    SELECT
        "Product Name" AS Product_Name,
        Sales
    FROM superstore
)
GROUP BY Product_Name
ORDER BY Ventas DESC
LIMIT 10;""")
```

Consulta 4. Ventas por segmento de cliente

```python
duckdb.sql("""SELECT
    Segment,
    COUNT(*) AS Cantidad_Pedidos,
    ROUND(SUM(Sales),2) AS Total_Ventas
FROM superstore
GROUP BY Segment
ORDER BY Total_Ventas DESC;""")
```

Consulta 5. Estados con mayor utilidad

```python
duckdb.sql("""SELECT
    State,
    ROUND(SUM(Profit),2) AS Utilidad_Total
FROM superstore
GROUP BY State
ORDER BY Utilidad_Total DESC
LIMIT 10;""")
```

Consulta 6. Descuento promedio por categoría

```python
duckdb.sql("""SELECT
    Category,
    ROUND(AVG(Discount),3) AS Descuento_Promedio
FROM superstore
GROUP BY Category;""")
```
